```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
t = pd.read_csv("1h-ckpt-celio.csv", names=["Bandwidth", "MTBF", "Interference", "WORK", "IO", "CKPT", "WASTED", "TOTAL", "seed"])
v = t.groupby(["Interference","seed"]).mean()
v.reset_index(inplace=True)
v['TOTWASTE']=v['CKPT']+v['WASTED']
v['TOTWORK']=v['WORK']+v['IO']
baseline = 1.0/v[v.Interference == "baseline"]['TOTWORK'].min()
v=v[v.Interference != "baseline"]
v.reset_index(inplace=True)
v['Computing Ratio']=v['WORK']*baseline
v['IO Ratio']=v['IO']*baseline
v['Checkpoint Ratio']=v['CKPT']*baseline
v['Wasted Computing Ratio']=v['WASTED']*baseline
v['Waste Ratio']=v['TOTWASTE']*baseline
v['Work Ratio']=v['TOTWORK']*baseline
print v.count()
print t.count()
```

```python
data=v[v.Bandwidth == 4e10]
figname="1hckpt-40Gbs-"

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='Computing Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig.savefig(figname + "computing-ratio.pdf")

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='IO Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
fig.savefig(figname + "io-ratio.pdf")

ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='Work Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig.savefig(figname + "work-ratio.pdf")

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='Wasted Computing Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig.savefig(figname + "wasted-computing-ratio.pdf")
             
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='Checkpoint Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig.savefig(figname + "checkpoint-ratio.pdf")

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.boxplot(x="MTBF", y='Waste Ratio', hue="Interference", 
            data=data, palette="colorblind", 
            ax=ax)
ax.set_title("Aggregated Bandwidth: 40GB/s\n1H Fixed Checkpoint Interval")
fig.savefig(figname + "waste-ratio.pdf")
```
